# Problem 1

(5 points)

Write a function to calculate a unit vector.

* The function should take 2 points as input, with each point containing the [x,y,z] coordinates.  
* The function should compute the displacement between the two points, and then divide the displacement vector by the length of the vector.

$$\hat{r} = \frac{\vec{r}}{\left|\vec{r}\right|} $$

* The function should return the unit vector.
* Note: you can calculate $\left|\vec{r}\right|$ in python using
```python
point1 = np.array([1,2,3])
norm_point1 = np.linalg.norm(point1)
```


````python
# Your answer here

def unitv(vector1, vector2):
  """
  Takes in two vectors as an input and returns a unit vector
  INPUT
  vector1: input vector
  vector2: input vector

  RETURNS
  unit_vector: normalized unit vector using the two input vectors
  """

  dvector = vector2 - vector1
  normv = np.linalg.norm(dvector)

  unit_vector = dvector / normv

  return unit_vector
````

````python
vector1 = np.array([1,2,3])
vector2 = np.array([2,4,6])

unitv(vector1,vector2)
````

* Using your function from Problem 1, compute the unit vector that points from $p1$ to $p2$.
* Draw the unit vector on your plot.  See https://matplotlib.org/3.4.3/gallery/text_labels_and_annotations/arrow_simple_demo.html for details on how to draw an arrow.

````python
# Your answer here
p1 = np.array([2,2,2]) # Point 1
p2 = np.array([5,5,5]) # Point 2
p3 = unitv(p1,p2) # Function to create unit vector just labeled as p3

plt.figure()
ax = plt.axes(projection='3d')

plt.plot(p1[0],p1[1],p1[2],'ro',markersize=12,label='p1') # Plot of point 1
plt.plot(p2[0],p2[1],p2[2],'bs',markersize=12,label='p2') # Plot of point 2

ax.quiver(p1[0],p1[1],p1[2], # Plot for the unit vector
          p3[0],p3[1],p3[2],
          color='g', label='Unit Vector')

ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')

ax.legend(loc='upper left')
````

# Problem 3

(10 points)

* Write a function that computes the first 20 terms of the Fibonnaci sequence.
* Make a plot of the Fibonnaci number versus its number in the sequence.

````python
# Your code here

def fibseq(terms):
  """
  This function takes in one variable (term) and returns the desired term of the fibonacci sequence.
  INPUT:
  terms: integer to determine to which term of the fibonacci sequence you want to go to
  """
  numlist=[0,1]
  for i in range(terms-1):
    num = numlist[i+1] + numlist[i]
    numlist.append(num)
  return num,numlist
````

````python
fibseq(20)[0]
indice = np.linspace(1,len(fibseq(20)[1]),len(fibseq(20)[1]))
````

````python
plt.plot(indice,fibseq(20)[1],'b.')
````
