# Write the initialize function

Write a function called **initialize** that does the following.
* Input
  * initial x position, $x_0$
  * initial y position, $y_0$
  * initial x velocity, $v_{0x}$
  * initial y velocity, $v_{0y}$
  * total time, $tmax$
  * time step, $dt$

* The function should:
  *  calculate the number of time steps
  ```
  nsteps = int(tmax/dt)
  ```
  * create the time, x, y, vx, vy arrays using `np.zeros`
  * set the first value of the x array to $x_0$
  * set the first value of the y array to $y_0$
  * set the first value of the vx array to $v_{0x}$
  * set the first value of the vy array to $v_{0y}$

* The function should return the following arrays *in this order*:
  * time array
  * x position array
  * y position array
  * x velocity array
  * y velocity array

````python
# write your initialize function here
def initialize (x0e, y0e, vx0e, vy0e, x0j, y0j, vx0j, vy0j, tmax, dt):
  """
  Sets up initial arrays of zeros for x and y position and velocity and time, also sets any initial values

  INPUTS:
  x0 - initial x position
  y0 - initial y position
  vx0 - initial x velocity
  vy0 - initial y velocity
  tmax - total time
  dt - time step

  RETURNS:
  time - time array
  x - x position array
  y - y position array
  vx - x velocity array
  vy - y velocity array
"""
  nsteps = int(tmax/dt)

  time = np.zeros(nsteps)
  xe = np.zeros(nsteps)
  ye = np.zeros(nsteps)
  vxe = np.zeros(nsteps)
  vye = np.zeros(nsteps)

  xj = np.zeros(nsteps)
  yj = np.zeros(nsteps)
  vxj = np.zeros(nsteps)
  vyj = np.zeros(nsteps)

  xe[0] = x0e
  ye[0] = y0e
  vxe[0] = vx0e
  vye[0] = vy0e

  xj[0] = x0j
  yj[0] = y0j
  vxj[0] = vx0j
  vyj[0] = vy0j

  return time, xe, ye, vxe, vye, xj, yj, vxj, vyj
````

# Distance Function

Write a function called **distance** to calculate the distance between the Earth and Sun given the Earth's x and y positions.  We assume throughout that the Sun is located at $x=0, y=0$, and that the Sun does not move.

Write a function called **distance** that does the following.
* Input
  * x position, $x$
  * y position, $y$

* The function should:
  *  calculate the distance between the Earth and Sun
  ```
  r = np.sqrt(x**2 + y**2)
  ```


* The function should return :
  * the distance $r$

````python
# write your distance function here
def distance(x, y):
  """
  Calculates distance array using x and y array
  """
  r = np.sqrt(x**2 + y**2)
  return r
````

# Calculate function

Write a function called **calculate** that does the following.
* Input
  * time array, $time$
  * x position, $x$
  * y position, $y$
  * x velocity, $v_{x}$
  * y velocity, $v_{y}$
  * time step, $dt$
  * optional argument **cromer** that has a default value of ```True```

* The function should:
  * calculate vx, vy, and time
  * if ```cromer=True```, use the Euler-Cromer method to calculate x and y
  * if ```cromer=False```, use the Euler method to calculate x and y

* The function should return the following arrays *in this order*:
  * time array
  * $x$ position array
  * $y$ position array
  * $v_{x}$ velocity array
  * $v_{y}$ velocity array

````python
# write your calculate function
# write your calculate function
def calculate(time, xe, ye, vxe, vye, xj, yj, vxj, vyj, me, mj, ms, dt, cromer=True):
  for i in range(1,len(time)):
    time[i] = time[i-1] + dt
    vxe[i] = vxe[i-1] + (((-4 * np.pi ** 2) / (distance(xe[i-1],ye[i-1])**3)) * xe[i-1]) * dt - (4 * np.pi**2 *(mj/ms)*(xe[i-1]-xj[i-1])/(distance(xj[i-1] - xe[i-1], yj[i-1] - ye[i-1]))**3)*dt
    vye[i] = vye[i-1] + (((-4 * np.pi ** 2) / (distance(xe[i-1],ye[i-1])**3)) * ye[i-1]) * dt - (4 * np.pi**2 *(mj/ms)*(ye[i-1]-yj[i-1])/(distance(xj[i-1] - xe[i-1], yj[i-1] - ye[i-1]))**3)*dt

    vxj[i] = vxj[i-1] + (((-4 * np.pi ** 2) / (distance(xj[i-1],yj[i-1])**3)) * xj[i-1]) * dt - (4 * np.pi**2 *(me/ms)*(xj[i-1]-xe[i-1])/(distance(xj[i-1] - xe[i-1], yj[i-1] - ye[i-1]))**3)*dt
    vyj[i] = vyj[i-1] + (((-4 * np.pi ** 2) / (distance(xj[i-1],yj[i-1])**3)) * yj[i-1]) * dt - (4 * np.pi**2 *(me/ms)*(yj[i-1]-ye[i-1])/(distance(xj[i-1] - xe[i-1], yj[i-1] - ye[i-1]))**3)*dt
    if cromer == True:
      xe[i] = xe[i-1] + vxe[i] * dt
      ye[i] = ye[i-1] + vye[i] * dt

      xj[i] = xj[i-1] + vxj[i] * dt
      yj[i] = yj[i-1] + vyj[i] * dt
    else:
      xe[i] = xe[i-1] + vxe[i-1] * dt
      ye[i] = ye[i-1] + vye[i-1] * dt

      xj[i] = xj[i-1] + vxj[i-1] * dt
      yj[i] = yj[i-1] + vyj[i-1] * dt
  return time, xe, ye, vxe, vye, xj, yj, vxj, vyj
````

# Function to make plots

Write a function called **make_plots** that does the following.
* Input
  * time array, $time$
  * x position, $x$

````python
# write your make_plots function here
def make_plots(time, xe, ye, vxe, vye, xj, yj, vxj, vyj):
  plt.figure(figsize=(16,8))
  plt.subplots_adjust(wspace=.35)

  # Orbital Plot
  plt.subplot(2,4,1)
  plt.scatter(xe,ye,c=time,s=10, label='Earth')
  plt.colorbar(label="Time (yr)")
  plt.axis('equal')
  plt.xlabel('Time (yr)')
  plt.ylabel('Position (au)')
  plt.legend()

  plt.subplot(2,4,5)
  plt.scatter(xj, yj, c=time, s=10, label='Jupiter')

  plt.colorbar(label="Time (yr)")
  plt.axis('equal')
  plt.xlabel('Time (yr)')
  plt.ylabel('Position (au)')
  plt.legend()

  plt.subplot(2,4,2)
  plt.scatter(time, xe, label='Earth X Position')
  plt.scatter(time, distance(xe,ye), label='Earth Radius')
  plt.xlabel('Time (yr)')
  plt.ylabel('Position (au)')
  plt.legend()

  plt.subplot(2,4,3)
  plt.scatter(time, ye, label='Earth Y Position')
  plt.scatter(time, distance(xe,ye), label='Earth Radius')
  plt.xlabel('Time (yr)')
  plt.ylabel('Position (au)')
  plt.legend()

  plt.subplot(2,4,6)
  plt.scatter(time, xj, label='Jupiter X Position')
  plt.scatter(time, distance(xj,yj), label='Jupiter Radius')
  plt.xlabel('Time (yr)')
  plt.ylabel('Position (au)')
  plt.legend()

  plt.subplot(2,4,7)
  plt.scatter(time, yj, label='Jupiter Y Position')
  plt.scatter(time, distance(xj,yj), label='Jupiter Radius')
  plt.xlabel('Time (yr)')
  plt.ylabel('Position (au)')
  plt.legend()

  plt.subplot(2,4,4)
  plt.scatter(time, vxe, label='Earth X Velocity')
  plt.scatter(time, vye, label='Earth Y Velocity')
  plt.scatter(time, np.sqrt(vxe**2 + vye**2), label='Earth Total Velocity')
  plt.xlabel('Time (yr)')
  plt.ylabel(r'Velocity ($\frac{au}{yr}$)')
  plt.legend()

  plt.subplot(2,4,8)
  plt.scatter(time, vxj, label='Jupiter X Velocity')
  plt.scatter(time, vyj, label='Jupiter Y Velocity')
  plt.scatter(time, np.sqrt(vxj**2 + vyj**2), label='Jupiter Total Velocity')
  plt.xlabel('Time (yr)')
  plt.ylabel(r'Velocity ($\frac{au}{yr}$)')
  plt.legend()
````

# main function

Write a function called **main** that does the following.  Note, this is different from how we have structured the main program in the past, but it will allow us to modify the parameters of the model easily.

* The function should return the following arrays *in this order*:
  * time array
  * $x$ position array

````python
# write your main function here
def main(x0e, y0e, vx0e, vy0e, x0j, y0j, vx0j, vy0j, tmax, dt, me, mj, ms, cromer=True):
  time, xe, ye, vxe, vye, xj, yj, vxj, vyj = initialize(x0e, y0e, vx0e, vy0e, x0j, y0j, vx0j, vy0j, tmax, dt)

  time, xe, ye, vxe, vye, xj, yj, vxj, vyj = calculate(time, xe, ye, vxe, vye, xj, yj, vxj, vyj, me, mj, ms, dt, cromer)

  make_plots(time, xe, ye, vxe, vye, xj, yj, vxj, vyj)

  return time, xe, ye, vxe, vye, xj, yj, vxj, vyj
````
